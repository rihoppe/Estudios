```python
datos = open('result.csv','r')
contenido = datos.readlines()

for i in range(len(contenido)):
	contenido[i] = contenido[i].split(',')
	contenido[i][8]=contenido[i][8][:-1]
```

```python
#toma y separaciÃ³n de TODOS los datos

#[id , ra, dec, Redshift, Redshift Err, Mag_g, Mag_r, Mag_g_Err, Mag_r_Err ]
redshift_1,redshift_2,redshift_3,redshift_4,redshift_5 = [],[],[],[],[]
color_1,color_2,color_3,color_4,color_5 = [],[],[],[],[]
ra_1, ra_2, ra_3, ra_4, ra_5 = [],[],[],[],[]
redshift_6,redshift_7,redshift_8,redshift_9,redshift_10 = [],[],[],[],[]
color_6,color_7,color_8,color_9,color_10 = [],[],[],[],[]
ra_6, ra_7, ra_8, ra_9, ra_10 = [],[],[],[],[]
ra,dec,redshift,redshift_err,mag_g,mag_r,mag_g_err,mag_r_err,color_tot,color_tot_err,ra_esp,redshift_esp = [],[],[],[],[],[],[],[],[],[],[],[]

for i in range(len(contenido)):
    if i!=0 and float(contenido[i][3])>=0:
            ra.append(float(contenido[i][1]))
            dec.append(float(contenido[i][2]))
            redshift.append(float(contenido[i][3]))
            redshift_err.append(float(contenido[i][4]))
            mag_g.append(float(contenido[i][5]))
            mag_r.append(float(contenido[i][6]))
            mag_g_err.append(float(contenido[i][7]))
            mag_r_err.append(float(contenido[i][8]))
for i in range(len(mag_g)):
        color_tot.append(float(mag_g[i])-float(mag_r[i]))
        color_tot_err.append(np.sqrt(float(mag_g_err[i])**2+float(mag_r_err[i])**2))
for i in range(len(redshift)):       
        if redshift[i]>0 and redshift[i]<=0.05:
                redshift_1.append(redshift[i])
                color_1.append(color_tot[i])
                ra_1.append(float(ra[i])*np.pi/180)
        elif redshift[i]>0.05 and redshift[i]<=0.1:
                redshift_2.append(redshift[i])
                color_2.append(color_tot[i])
                ra_2.append(float(ra[i])*np.pi/180)
        elif redshift[i]>0.1 and redshift[i]<=0.15:
                redshift_3.append(redshift[i])
                color_3.append(color_tot[i])
                ra_3.append(float(ra[i])*np.pi/180)
        elif redshift[i]>0.15 and redshift[i]<=0.20:
                redshift_4.append(redshift[i])
                color_4.append(color_tot[i])
                ra_4.append(float(ra[i])*np.pi/180)
        elif redshift[i]>0.20 and redshift[i]<=0.25:
                redshift_5.append(redshift[i])
                color_5.append(color_tot[i])
                ra_5.append(float(ra[i])*np.pi/180)
        elif redshift[i]>0.25 and redshift[i]<=0.30:
                redshift_6.append(redshift[i])
                color_6.append(color_tot[i])
                ra_6.append(float(ra[i])*np.pi/180)
        elif redshift[i]>0.30 and redshift[i]<=0.35:
                redshift_7.append(redshift[i])
                color_7.append(color_tot[i])
                ra_7.append(float(ra[i])*np.pi/180)
        elif redshift[i]>0.35 and redshift[i]<=0.40:
                redshift_8.append(redshift[i])
                color_8.append(color_tot[i])
                ra_8.append(float(ra[i])*np.pi/180)
        elif redshift[i]>0.40 and redshift[i]<=0.45:
                redshift_9.append(redshift[i])
                color_9.append(color_tot[i])
                ra_9.append(float(ra[i])*np.pi/180)
        elif redshift[i]>0.45 and redshift[i]<=0.5:
                redshift_10.append(redshift[i])
                color_10.append(color_tot[i])
                ra_10.append(float(ra[i])*np.pi/180)
        
        
for i in range(len(ra)):
    ra[i]=float(ra[i])*np.pi/180
    dec[i]=float(dec[i])*np.pi/180
    if float(ra[i])>=np.pi*0.6 and float(ra[i])<=3*np.pi/2:
        ra_esp.append(ra[i])
        redshift_esp.append(redshift[i])
```

```python
from matplotlib.transforms import Affine2D
import mpl_toolkits.axisartist.floating_axes as floating_axes
import numpy as np
import mpl_toolkits.axisartist.angle_helper as angle_helper
from matplotlib.projections import PolarAxes
from mpl_toolkits.axisartist.grid_finder import (FixedLocator, MaxNLocator,
                                                 DictFormatter)
import matplotlib.pyplot as plt

def setup_axes2(fig, rect,r0,r1,theta0,theta1):
    """
    With custom locator and formatter.
    Note that the extreme values are swapped.
    """
    tr_rotate = Affine2D().rotate(3*np.pi/2)
    tr_scale = Affine2D().scale(2, 1)
    tr = PolarAxes.PolarTransform() + tr_rotate #+ tr_scale +  Affine2D().expanded(2, 2)
    pi = np.pi
    angle_ticks = [(0, r"$0$"),(pi*1.5, r"$\frac{3}{4}\pi$"),(pi, r"$\pi$"),
                  (pi/2,r"$\frac{1}{2}\pi$"),
                   (2*pi/3,r"$\frac{2}{3}\pi$"),
                   (4*pi/3,r"$\frac{4}{3}\pi$"), (5*pi/6,r"$\frac{5}{6}\pi$"),
                  (7*pi/6,r"$\frac{5}{6}\pi$")]
    
    grid_locator1 = FixedLocator([v for v, s in angle_ticks])
    #grid_locator1 = MaxNLocator(10)
    tick_formatter1 = DictFormatter(dict(angle_ticks))
    grid_locator2 = MaxNLocator(5)
#1.47*pi, 0.55*pi
    grid_helper = floating_axes.GridHelperCurveLinear(
        tr, extremes=(theta1*pi,theta0*pi, r1, r0),
        grid_locator1=grid_locator1,
        grid_locator2=grid_locator2,
        tick_formatter2=None,
        tick_formatter1=tick_formatter1)

    ax1 = floating_axes.FloatingSubplot(fig, rect, grid_helper=grid_helper)
    fig.add_subplot(ax1)

    # create a parasite axes whose transData in RA, cz
    aux_ax = ax1.get_aux_axes(tr)

    aux_ax.patch = ax1.patch  # for aux_ax to have a clip path as in ax
    ax1.patch.zorder = 0.9  # but this has a side effect that the patch is
    # drawn twice, and possibly over some other
    # artists. So, we decrease the zorder a bit to
    # prevent this.

    return ax1, aux_ax
```

```python
%matplotlib inline
fig = plt.figure(num=1,dpi=300)
ax3, aux_ax3 = setup_axes2(fig, 111,0,1,0.55,1.47)

#ax3.axis["bottom"].label.set(text="something", color="red",
#                            family="serif", size=30)

aux_ax3.scatter(ra, redshift,s=1)
ax3.grid(True)
plt.savefig("medio_pie_grande.pdf")
```
