```python
import numpy as np
import scipy
from scipy.spatial import distance
import scipy.constants as constants
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def ob_m_ap(m_abs,dist): #mega-parsecs
	return m_abs-5+(5*np.log10(dist*1000000))
def ob_z(dist): #mega-parcecs
	c = 3*(10**5)
	h = 70.
	return dist*h/(c)
```

```python
m_abs = []
x = []
y = []
z = []
for i in range(4000):
	m_abs.append(random.uniform(-22,-19))
	x.append(random.uniform(-3000,3000))
	y.append(random.uniform(-3000,3000))
	z.append(random.uniform(-3000,3000))
print len(x),len(y),len(z),len(m_abs)
#print m_ap
```

```python
m_abs_cir = []
x_cir = []
y_cir = []
z_cir = []
for i in range(4000):
    x_c = random.uniform(-3000,3000)
    y_c = random.uniform(-3000,3000)
    z_c = random.uniform(-3000,3000)
    if ((x_c**2)+(y_c**2)+(z_c**2))<=9000000:
        x_cir.append(x_c)
        y_cir.append(y_c)
        z_cir.append(z_c)
for i in range(len(x_cir)):
    m_abs_cir.append(random.uniform(-22,-19))
print len(x_cir),len(y_cir),len(z_cir),len(m_abs_cir)
```

```python
dist = []
for i in range(len(x)):
    l = (0,0,0)
    k = (x[i],y[i],z[i])
    dist.append(distance.euclidean(l,k))
print len(dist)
m_ap = []
for i in range(len(m_abs)):
    m_ap.append(ob_m_ap(m_abs[i],dist[i]))
theta = []
for i in range(len(x)):
    a = np.arctan2(y[i],x[i])
    if a<0:
        a+=2*np.pi
    theta.append(a)
#print theta
#print dist
```

```python
dist_cir = []
for i in range(len(x_cir)):
    l = (0,0,0)
    k = (x_cir[i],y_cir[i],z_cir[i])
    dist_cir.append(distance.euclidean(l,k))
print len(dist_cir)
m_ap_cir = []
for i in range(len(m_abs_cir)):
    m_ap_cir.append(ob_m_ap(m_abs_cir[i],dist_cir[i]))
theta_cir = []
for i in range(len(x_cir)):
    a = np.arctan2(y_cir[i],x_cir[i])
    if a<0:
        a+=2*np.pi
    theta_cir.append(a)
#print theta
#print dist
```

```python
%matplotlib inline
print len(theta),len(dist)
ax = plt.subplot(111, projection='polar')
ax.scatter(theta, dist, s=5)
ax.grid(True)
plt.show()
```

```python
redshift = []
redshift_cir =[]
for i in range(len(dist)):
    redshift.append(ob_z(dist[i]))
for i in range(len(dist_cir)):
    redshift_cir.append(ob_z(dist_cir[i]))
```
